```python
# Project that generates a report showing what users are currently logged into a server. 
# Analyzes a dictionary input containing time, event (login or logout), server name, and username.

def get_event_date(event):
	return event.date

def current_users(events):
	events.sort(key=get_event_date)
	machines={}
	for event in events:
		if event.machine not in machines:
			machines[event.machine]=set()
		if event.type == "login":
			machines[event.machine].add(event.user)
		elif event.type == "logout":
			machines[event.machine].remove(event.user)
	return machines

def generate_report(machines):
	for machine, users in machines.items():
		if len(users) > 0:
			user_list=", ".join(users)
			print("{}: {}".format(machine, user_list))
```

```python
# To use this code, we will use the Event class
class Event:
    def __init__(self, event_date, event_type, machine_name, user):
        self.date = event_date
        self.type = event_type
        self.machine = machine_name
        self.user = user
```

```python
# Sample events
events = [
    Event('2020-01-21 12:45:56', 'login', 'myworkstation.local', 'Jordan'),
    Event('2020-01-22 15:53:42', 'logout', 'webserver.local', 'Jordan'),
    Event('2020-01-21 18:53:21', 'login', 'webserver.local', 'Lane'),
    Event('2020-01-22 10:25:34', 'logout', 'myworkstation.local', 'Jordan'),
    Event('2020-01-21 08:20:01', 'login', 'webserver.local', 'Jordan'),
    Event('2020-01-23 11:24:35', 'login', 'mailserver.local', 'Chris'),
]
```

```python
# Let's call the code and verify that it's functional
users = current_users(events)
print(users)
```

```python
# Generate the report
report=generate_report(users)
```
